```python
# import 模組
from pathlib import Path
import pandas as pd
from ydata_profiling import ProfileReport
import ipywidgets as widgets

# 路徑設定
projRoot = Path.cwd()
root_st1data = projRoot.joinpath('data/stage1')
```

### Step1. 選取資料類別

請先執行下方語法，並選取要分析資料的類別

```python
filepaths = sorted(root_st1data.rglob('*.csv'))

options_cate = list(set(fp.name[0] for fp in filepaths)) # 保留單一值

cate_togl =  widgets.ToggleButtons(
    options=options_cate,
    description='請選取資料表類別:',
    disabled=False
)
cate_togl
```

### Step3. 選取要檢視的資料表

執行完下方語法後，由下拉選單中選取要分析的資料表。

若重新點選上方的類別，需重新執行下方語法。

```python
options_lt = [fp.name for fp in sorted(root_st1data.rglob(f'{cate_togl.value}*'))]

dropdwn_dt = widgets.Dropdown(
    options=options_lt,
    # value='',
    description='請點選資料表:'
    # disabled=False,
)
dropdwn_dt
```

##### 欄位簡介

```python
path_man = projRoot.joinpath(f'man/raw/{dropdwn_dt.value}')
dtman =  pd.read_csv(path_man)
dtman
```

##### 描述統計報表

```python
df = pd.read_csv(root_st1data.joinpath(dropdwn_dt.value),
                 dtype={'學校統計處代碼': 'string', '系所代碼': 'string'}
                 )
profile = ProfileReport(df, title="Profiling Report", minimal=True)
# 輸出速度有影響
print(f'{dropdwn_dt.value}')
profile.to_notebook_iframe()
```
